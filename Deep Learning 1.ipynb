## Types of Deep Learning Models

ANN (Artificial Neural Network): Tabular / Structured Data

CNN (Convolutional Neural Network): Image-related data

RNN (Recurrent Neural Network): Sequential data (text, time-series)

LSTM: Handles long-term dependencies

GRU: Similar to LSTM but faster

### Activation Functions

Purpose: Convert linear input into non-linear representation so the network can learn complex patterns.

Sigmoid: Output between 0 & 1 (for binary classification)

Tanh: Output between -1 & 1

ReLU (Rectified Linear Unit): Most common for hidden layers

Softmax: For multi-class classification

# Optimizers

SGD (Stochastic Gradient Descent)

Adam (Adaptive Moment Estimation)

RMSprop

### Loss Functions

Regression: MSE (Mean Squared Error)

Binary Classification: Binary Cross-Entropy

Multi-Class Classification: Categorical Cross-Entropy
